## Import libraries ##

```python
import pandas as pd   
import altair as alt
```

```python
data = pd.read_csv("global_firepower.csv")
data.shape
```

```python
data.head()
```

```python
data.columns = data.columns.str.replace(' ', '_').str.lower()
data.head()
```

## 5 countries with the most Number of Attack Helicopter ##

```python
attack_helicop_num = data[data["attack_helicopters"]>100]
attack_helicop_num = attack_helicop_num[["country", "attack_helicopters"]].sort_values(by="attack_helicopters", ascending=False).reset_index()
attack_helicop_num
```

```python
chart= (alt.Chart(attack_helicop_num)
            .mark_bar()
            .encode(x = 'country', y="attack_helicopters").properties(width = 800))
chart.save("atack_helicop_num.png")
chart
```

## Countries with more than 1200 tanks ##

```python
tanks = data[data["tanks"]>1200]
tanks = tanks[["country", "tanks"]].sort_values(by="tanks", ascending=False)
tanks
```

```python
chart= (alt.Chart(tanks)
            .mark_bar()
            .encode(x = 'country', y="tanks").properties(width = 800))
chart.save("tanks.png")
chart
```

## Proportion of active personnel and population ##

```python
proportion = data[["country", "active_personnel","total_population"]]
proportion = proportion.assign(proportion = lambda x: (x.active_personnel*100/x.total_population).round(2))
proportion = proportion[proportion["proportion"]>1].sort_values(by="proportion", ascending = False)
proportion
```

```python
chart= (alt.Chart(proportion)
            .mark_bar()
            .encode(x = 'country', y="proportion").properties(width = 800))
chart.save("proportion.png")
chart
```
