````python
import pandas as pd   
import numpy as np
import altair as alt
from altair_saver import save
````

````python
url ="https://raw.githubusercontent.com/byuidatascience/data4missing/master/data-raw/flights_missing/flights_missing.json"
url= pd.read_json(url)
data=url
````

````python
question1= (url.groupby("airport_code")
                .agg(total_flights = ("num_of_flights_total", sum),
                    total_delays = ("num_of_delays_total", sum),
                    total_delay_min = ("minutes_delayed_total",sum))
                .assign(prop_delays = lambda x: (x.total_delays/x.total_flights*100).round(2),
                        avg_delay_in__hours = lambda x: x.total_delay_min/x.total_delays/60
                        )
                .reset_index()
            )
print(question1.to_markdown(index=False))
````

````python
chart= (alt.Chart(question1)
            .mark_bar()
            .encode(x = 'airport_code', y="total_delays").properties(width = 800))
chart.save("asdasd.png")
chart
````

````python

question2 = url.query("month != 'n/a'")
question2 = (question2.groupby("month")
                .agg(total_flights = ("num_of_flights_total", sum),
                    total_delays = ("num_of_delays_total", sum),
                    )
                .assign(Porcentage_of_delay = lambda x: (x.total_delays/x.total_flights*100).round(2),
                        )
                .reset_index()
            )
question2
````

````python
ordered_months = ["January", "Febuary", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
ordered_months
````

````python
chart= (alt.Chart(question2)
            .mark_bar()
            .encode(x = alt.X('month', sort= ordered_months), y="Porcentage_of_delay").properties(width = 800))
chart.save("proyect.png")
chart
````

````python
#url.replace(-999, np.nan)
q3 = url
q3.num_of_delays_late_aircraft.replace(-999, np.nan, inplace=True)
bob = q3.num_of_delays_late_aircraft.mean()
q3.num_of_delays_late_aircraft.replace(np.nan, bob, inplace=True)
#q3.num_of_delays_late_aircraft.fillna(value =  bob, inplace=True)
q3.head()
````

````python
weather = q3.assign(
    severe = q3.num_of_delays_weather,
    mild_late = .3*q3.num_of_delays_late_aircraft.round(2),
    mild_nas = np.where(q3.month.isin(["April", "May", "June", "July", "August"]),.4*q3.num_of_delays_nas,.65*q3.num_of_delays_nas),
    total_delay_due_to_weather= lambda x: (x.severe + x.mild_late + x.mild_nas).round(2)
)
print(weather.head(5).filter(["airport_code", "severe", "mild_late", "mild_nas", "total_delay_due_to_weather"]).to_markdown(index=False))
````

````python
q4 = (weather.groupby("airport_code")
        .agg(total_flights=("num_of_flights_total", sum),
             total_delay_due_to_weather = ("total_delay_due_to_weather", sum))
            .assign(prop_delay_by_weather = lambda x: (x.total_delay_due_to_weather/x.total_flights*100).round(2))
                .reset_index()
)
chart= (alt.Chart(q4)
            .mark_bar()
            .encode(x = "airport_code", y="prop_delay_by_weather").properties(width = 800))
#chart.save("proyect2.2.png")
chart
````
